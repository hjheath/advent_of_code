```python
test = """
.....
.S-7.
.|.|.
.L-J.
.....
"""

test2 = """
..F7.
.FJ|.
SJ.L7
|F--J
LJ...
"""

# grid = [line for line in test2.strip().splitlines()]

file_path = 'day_10_input.txt'
with open(file_path, 'r') as file:
    grid = [line for line in file]

def get_start_coords(grid):
    for y, row in enumerate(grid):
        for x, col in enumerate(row):
            if col == 'S':
                return x, y
    return None

def get_start_connections(grid, start_coords):
    connected_pipes = []
    permitted = {
        'right': ['-', 'J', '7'],
        'left': ['-', 'L', 'F'],
        'up': ['|', '7', 'F'],
        'down': ['|', 'L', 'J']
    }

    x, y = start_coords
    directions = {'right': (1, 0), 'left': (-1, 0), 'down': (0, 1), 'up': (0, -1)}

    for direction, coords in directions.items():
        i, j = coords
        x_next = x + i
        y_next = y + j
        
        # Coords out of range
        if x_next < 0 or y_next < 0 or x_next >= len(grid[0]) or y_next >= len(grid):
            pass
        element = grid[y_next][x_next]
        if element in permitted[direction]:
            connected_pipes.append((x_next, y_next))

    return connected_pipes

def apply_translation(current_pipe, translation):
    x = current_pipe[0] + translation[0]
    y = current_pipe[1] + translation[1]
    return x, y

def next_pipe(grid, current_pipe, prev_pipe):
    translations_map = {
        '-': [(1, 0), (-1, 0)],
        '|': [(0, 1), (0, -1)],
        'L': [(0, -1), (1, 0)],
        'J': [(0, -1), (-1, 0)],
        '7': [(0, 1), (-1, 0)],
        'F': [(0, 1), (1, 0)],
    }
    x, y = current_pipe
    element = grid[y][x]

    translations = translations_map[element]
    connections = [apply_translation(current_pipe, translation) for translation in translations]
    connections.remove(prev_pipe)
    return connections[0]

start_coords = get_start_coords(grid)
current_coords, _ = get_start_connections(grid, start_coords)
prev_coords = start_coords

counter = 1

while current_coords != start_coords:
    counter += 1
    next_coords = next_pipe(grid, current_coords, prev_coords)
    prev_coords, current_coords = current_coords, next_coords

counter // 2
```
