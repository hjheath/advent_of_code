```python
test = """
???.### 1,1,3
.??..??...?##. 1,1,3
?#?#?#?#?#?#?#? 1,3,1,6
????.#...#... 4,1,1
????.######..#####. 1,6,5
?###???????? 3,2,1
"""

def parse_line(line):
    report, meta = line.split(' ')
    meta = list(map(int, meta.split(',')))
    return report, meta

def parse_meta(meta):
    output = ['.*']
    for i, count in enumerate(meta):
        for _ in range(count):
            output.append('#')
        if i != len(meta) - 1:
            output.append('.+')
    output.append('.*')
    return output

class DFA:
    def __init__(self, input):
        self.states = self.parse_input(input)

    # def parse_input2(self, input):
    #     state_count = len(input)
    #     states = []
    #     for index, entry in enumerate(input):
    #         if entry == '.*':
    #             if index < state_count - 1:
    #                 states.append({'.': 0, '#': 1})
    #             else:
    #                 states.append({'.': 0})
    #         elif entry ==  '#' and input[index - 1] == '#':
    #             states.append({'#': 1})
    #         elif entry == '.+':
    #             states.append({'.': 1})
    #             states.append({'.': 0, '#': 1})

    #     states_with_index = [{'index': index, **state} for index, state in enumerate(states)]

    #     for state in states_with_index:
    #         for key, value in state.items():
    #             if key == 'index':
    #                 continue
    #             index = state['index']
    #             state[key] = states_with_index[index + value]

    #     return states_with_index

    def parse_input(self, input):
        states = []
        for index, entry in enumerate(input):
            if entry == '.*':
                states.append({'.': 0} if index == len(input) - 1 else {'.': 0, '#': 1})
            elif entry == '#' and input[index - 1] == '#':
                states.append({'#': 1})
            elif entry == '.+':
                states.extend([{'.': 1}, {'.': 0, '#': 1}])

        states_with_index = [{'index': index, **state} for index, state in enumerate(states)]

        for index, state in enumerate(states_with_index):
            for key, value in state.items():
                if key != 'index':
                    state[key] = states_with_index[index + value]
    
        return states_with_index


from collections import Counter

def calculate_possible_arrangements(input, dfa):
    counter = Counter({0: 1})
    
    for char in input:
        new_counter = Counter()
        for state_i, count in counter.items():
            state = dfa.states[state_i]
            if (char == '.' or char == '?') and state.get('.'):
                new_state = state['.']
                new_counter[new_state['index']] += count
            if (char == '#' or char == '?') and state.get('#'):
                new_state = state['#']
                new_counter[new_state['index']] += count
        counter = new_counter

    return counter[len(dfa.states) - 1]

# data = [parse_line(line) for line in test.strip().splitlines()]

file_path = 'day_12_input.txt'
with open(file_path, 'r') as file:
    data = [parse_line(line) for line in file]

counts = []
for line in data:
    report, meta = line
    dfa = DFA(parse_meta(meta))
    combination_count = calculate_possible_arrangements(report, dfa)
    counts.append(combination_count)

sum(counts)
```
